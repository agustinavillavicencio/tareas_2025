# Problema 3
Acá, lo primero que creí había que hacer es establecer los valores, los que pueden ser cualquiera con la función del input. Después pedí que, a partir de la sumatoria de los valores, me mostrara el resultado y si es menor a 200 con el if, elif para saber si el numero es mayor o igual a 200 pero menor a 250 y else para cualquiera de los otros resultados.

```python

n1 = int(input("Ingresar el primer número: "))
n2 = int(input("Ingresar el segundo número: "))

resultado = n1 + n2
print("La suma es:", resultado)
if resultado < 200:
    print("menor a 200")
elif 200 <= resultado < 250:
    print("mayor a 200")
else:
    print("mayor a 250")
```

# Ejercicio 4

En este caso, primero definí que la evaluación del usuario sería por nombre, edad y si le gusta la música urbana, esos son los datos que hay que recopilar. Despúes, con los datos que se recolectaron/van a recolectar, pedí lo requerido. Lo que hice fue agregar el apartado que recolecte las respuestas que digan que alguien es menor de edad y que si le guste la música urbana o no. Y después, agregué lo definido para que me entregara la respuesta de esa manera (nombre, edad, y si le gusta la música urbana).

```python
def evaluar_usuario(nombre, edad, gusta_urbana):
    print(f"{nombre}")

    if edad >= 18:
        if gusta_urbana:
            print("Eres mayor de edad y te gusta la música urbana")
        else:
            print("Eres mayor de edad, no te gusta la música urbana")
    else:
        if gusta_urbana:
            print("Eres menor de edad y te gusta la música urbana")
        else:
            print("Eres menor de edad y no te gusta la música urbana")

nombre = input(" nombre ")
edad = int(input(" edad "))
respuesta_musica = input("¿Te gusta la música urbana? (si/no): ").strip()
gusta_urbana = True if respuesta_musica == "si" else False

evaluar_usuario (nombre, edad, gusta_urbana)
```
